# Integral and Weight

```python
from veroku.factors.experimental.truncated_factor import Truncated1DFactor
g1 = Gaussian(var_names=['a'], mean=0, cov=1, log_weight=0)
g2 = Gaussian(var_names=['a'], mean=2, cov=0.2, log_weight=0)
tf1 = Truncated1DFactor(non_truncated_factor=g1,
                  support_bounds=(0,4),
                  normalize=False)

tf2 = Truncated1DFactor(non_truncated_factor=g2,
                        support_bounds=(1.9,3),
                        normalize=False)
#assert raises NonOverLappingSupportsError
tf2.get_integral()
tf2.weight
```

```python
from veroku.factors.experimental.conditional_factor import ConditionalFactor
```

```python
factor_table = {(0,):g1,
                (1,):g2}
cf = ConditionalFactor(conditioning_var_names=["c"],
                       cardinalities=[2],
                       factor_table=factor_table)
```
